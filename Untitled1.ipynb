```python
# ode.py: a simple ODE solver (PED 2016, updated to python3 2019,
# minor changes 2023 and 2024)
#
# The program sets up a function solveODE(f,x0,tmax,dt) 
# which uses the Euler method to solve 
#  dx/dt = f(x,t)
# for x(t) and plot the result, with time running 
# from t = t0 = 0 to t = tmax, x(0) = x0, time-step = dt,
# and with f(x,t) specified by the string fSpec.

import numpy as np
import matplotlib.pyplot as plt

plt.ion()

def solveODE(fSpec,x0,tmax,dt):
       f = eval("lambda x,t:" + fSpec)
       t,x = 0.0,x0
       xk = [x0]
       tk = [0.0]
       plt.figure(1)
        while t < tmax - dt/2 :
           x += dt*f(x,t)
           t += dt
           tk.append(t)
           xk.append(x)
       plt.plot(tk,xk,'r.',markersize=2)
       plt.xlim(0.0,tmax*1.05)
       plt.show()
       plt.pause(0.001)
       print("Number of steps  = ",len(tk)-1)
       print("Final value of t = ",tk[-1])
       print("Final value of x = ",xk[-1])

print("\n\
 Enter ODE to solve by typing solveODE(\"f\",x0,tmax,dt) where \"f\"\n\
 is a string giving the function f(x,t) in the ODE dx/dt = f(x,t)")

```

```python
solveODE("x",1,1,0.1)
```

```python
solveODE('x',1,1,0.001)
```

```python
# plotexamples.py (PED 2018, 2022, 2024)
#
# very simple examples of plotting functions using numpy and mathplotlib

import numpy as np
import matplotlib.pyplot as plt

# set up a range for t from -10 to 10 with 401 points: 
t = np.linspace(-10,10,401)

# let's put the plot in figure 1:
plt.figure(1)

# plot the graph of t^2:
plt.plot(t,t*t)
plt.show()

# add in labels on axes; note use of latex inside $...$ for mathematical formulae:
plt.plot(t,t*t)
plt.xlabel("time $t$")
plt.ylabel("$t^2$")
plt.show()

# repeat but with a non-default range for the y axis:
plt.plot(t,t*t)
plt.ylim([-1,50])
plt.xlabel("time $t$")
plt.ylabel("$t^2$")
plt.show()

# repeat but with non-default ranges for the x and y axes:
plt.plot(t,t*t)
plt.xlim([-5,15])
plt.ylim([-1,50])
plt.xlabel("time $t$")
plt.ylabel("$t^2$")
plt.show()

# change the range of t:
t = np.linspace(0,20,401)

# plot the graph of sin(t):
plt.plot(t,np.sin(t))
plt.xlabel("time $t$")
plt.ylabel("$\sin(t)$")
plt.show()

# plot the graph of t*sin(t):
plt.plot(t,t*np.sin(t))
plt.xlabel("time $t$")
plt.ylabel("$t\,\sin(t)$")
plt.show()
```

```python
# ode2.py: a simple 2 dimensional ODE solver (PED 2016, updated to
# python3 2019, equal axes problem fixed 2022, minor changes 2023,
# 2024)
# This one is for easy use in jupyter so only plots at the end.
#
# The program sets up a function solveODE2(fSpec,gSpec,x0,y0,tmax,dt)
# which uses the Euler method to solve the 2-dimensional system
#  dx/dt = f(x,y,t), dy/dt = g(x,y,t)
# for x(t) and y(t) and plot the result, in the (x,y) plane, with time 
# from t = 0 to t = tmax, x(0) = x0, y(0) = y0, time-step = dt,
# and with f and g given by the strings fSpec and gSpec.
#
# example call:
# solveODE2('y','-x',1,0,2*np.pi,0.1)

import numpy as np
import matplotlib.pyplot as plt

plt.ion()

def solveODE2(fSpec,gSpec,x0,y0,tmax,dt):
       f = eval("lambda x,y,t:" + fSpec)
       g = eval("lambda x,y,t:" + gSpec)
       t,x,y = 0.0,x0,y0
       xk = [x0]
       yk = [y0]
       tk = [0.0]
       plt.figure(1)
       ax = plt.gca()
       ax.set_aspect('equal')
       plt.xlim(-1.45,1.45)
       plt.ylim(-1.45,1.45)
       plt.plot(xk,yk,'r.',markersize=8)
       while t < tmax - dt/2 :
           x,y = x+dt*f(x,y,t),y+dt*g(x,y,t)
           t += dt
           tk.append(t)
           xk.append(x)
           yk.append(y)
       plt.plot(xk,yk,'r.',markersize=4)
       plt.show()
       plt.pause(0.01)

       print("Number of steps  = ",len(tk)-1)
       print("Final value of t = ",tk[-1])
       print("Final value of x = ",xk[-1])
       print("Final value of y = ",yk[-1])
       print("Final value of x^2+y^2 = ",xk[-1]**2+yk[-1]**2)

print("\n\
 Enter ODE to solve by typing solveODE2(\"f\",\"g\",x0,y0,tmax,dt) where\n\
 \"f\" and \"g\" are strings giving the functions f(x,y,t) and g(x,y,t)\n\
 in the system of ODEs dx/dt = f(x,y,t), dy/dt = g(x,y,t)")
```

```python
solveODE2("y","-2*x",1,0,2*np.pi,0.001)
```
